# Exploratory Data Analysis

```python
# Create functions to load data

def get_date(col):
    """Pull the date from a string of date-time data"""
    
    return pd.to_datetime(col).dt.date


def get_time(col):
    """Pull the time from a string of date-time data"""

    return pd.to_datetime(col).dt.time


def wrangle(filepath, col1_date, col1, col2_date, col2, col3_date, col3):
    """Read in csv and create new columns for date and time as datetime objects
    from a string containing date-time data"""

    df = pd.read_csv(filepath)

    df[col1_date] = get_date(df[col1])
    df[col1] = get_time(df[col1])

    df[col2_date] = get_date(df[col2])
    df[col2] = get_time(df[col2])

    df[col3_date] = get_date(df[col3])
    df[col3] = get_time(df[col3])


    return df
```

```python
# Load data

clients = pd.read_csv('Clients-Grid.csv')
fleet = pd.read_csv('Fleet-Grid.csv')
requests = wrangle('Requests-Grid.csv',
                   'expected_dropff_date',
                   'expected_dropoff_time',
                   'pickup_date',
                   'pickup_time',
                   'dropoff_date',
                   'dropoff_time')
```

# Analysis

## How accurate is the new routing model?

The routing for the completed deliveries during the week of 6/10/22 to 6/16/22 does not appear to provide an accurate estimate for the delivery times. Of the 243 completed deliveries, approximately 84% took longer than the estimated delivery time. This could be the result of several factors, including an inaccurate routing model. If the distance between the pickup and dropoff facilities is represented by Euclidean distance, the model will consistently underestimate the distance, and consequently the time, between the two points.

Not only were the majority of deliveries late, but there were also packages that were not delivered. Across three separate days, (6/11, 6/13, and 6/14), there were 7 "missed deliveries." These packages do not have a dropoff time or an assigned courier. It is unclear whether the packages were ever assigned to a courier or not. Regardless, this is unacceptable for a company delivering medical supplies and must not happen again.

### Future Analysis
Further analysis could involve a more thorough examination of the routes themselves. With the provided data, it is unclear how the estimated dropoff times are determined. A more accurate model may include speed limits, routes along roads, current traffic patterns, etc.

The distribution of deliveries among couriers is another area to explore in the future. A more even workload across couriers could potentially prevent missed deliveries.
